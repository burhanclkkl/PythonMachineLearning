```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("polynomial_regression.csv", sep=";")
df.head()
```

```python
y = df.araba_max_hiz.values.reshape(-1,1)
x = df.araba_fiyat.values.reshape(-1,1)
```

```python
plt.scatter(x,y)
plt.xlabel("araba_max_hiz")
plt.ylabel("araba_fiyat")
plt.show()
#linear regression y=b0 + b1*x
#multiple linear regression y=b0 + b1*x1 + b2*x2
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x,y)
lr.predict(x)
```

```python
y_head = lr.predict(x)
print(y_head)
```

```python
plt.scatter(x,y)
plt.xlabel("araba_max_hiz")
plt.ylabel("araba_fiyat")

plt.plot(x,y_head, c="r")
plt.show()
```

```python
lr.predict([[10000]])
```

Simple Linear Regression : y = b0 + (b1 * x)
Polynomial Linear Regression: y = b0 + (b1 * x1) + (b2 * (x2^2) + ... + (bn * (xn * n))

rnek:

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_regression = PolynomialFeatures(degree=4) #degree = n (degree arttka doruluk pay artar)

x_polynomial = polynomial_regression.fit_transform(x) # polynomial' kullan uygula ve evir
print(x_polynomial)
```

```python
#fit
linear_regression2 = LinearRegression()
linear_regression2.fit(x_polynomial, y)

#gorsellestirme
#linear gorsellestirme
plt.scatter(x,y)
plt.xlabel("araba_max_hiz")
plt.ylabel("araba_fiyat")
plt.plot(x,y_head, c="r", label="linear")

#poly gorsellestirme
y_head2 = linear_regression2.predict(x_polynomial)
plt.plot(x, y_head2, color="g", label="poly")
plt.legend()
plt.show()
```
