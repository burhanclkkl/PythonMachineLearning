```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.random.normal(25,5,50)
y1 = np.random.normal(25,5,50)

x2 = np.random.normal(55,5,50)
y2 = np.random.normal(60,5,50)

x3 = np.random.normal(55,5,50)
y3 = np.random.normal(10,5,50)

x = np.concatenate((x1,x2,x3), axis = 0)
y = np.concatenate((y1,y2,y3), axis = 0)

dictionary = {"x":x, "y":y}
data = pd.DataFrame(dictionary)

plt.scatter(x1,y1, color = 'r', marker='.', alpha=0.3)
plt.scatter(x2,y2, color = 'b', marker='.', alpha=0.3)
plt.scatter(x3,y3, color = 'g', marker='.', alpha=0.3)
plt.show()
```

```python
#dendogram
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
merg = linkage(data, method="ward")
dendrogram(merg, leaf_rotation = 90)
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hiyerartical_cluster = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage="ward")
```

```python
cluster = hiyerartical_cluster.fit_predict(data)
```

```python
data["label"] = cluster
```

```python
plt.scatter(data.x[data.label == 0],data.y[data.label == 0], color = 'r', marker='.', alpha=0.3)
plt.scatter(data.x[data.label == 1],data.y[data.label == 1], color = 'b', marker='.', alpha=0.3)
plt.scatter(data.x[data.label == 2],data.y[data.label == 2], color = 'g', marker='.', alpha=0.3)
plt.show()
```
