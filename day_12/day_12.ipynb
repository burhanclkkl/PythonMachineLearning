```python
y = data.diagnosis.values
x_data = data.drop(["diagnosis"], axis=1)
```

```python
x_data
```

```python
y
```

```python
y = [1 if each == 'M' else 0 for each in data.diagnosis]
```

```python
y
```

```python
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data)) # normalization
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42) # train test split
```

```python
#knn model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 11) #n_neighbors = k
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
```

```python
prediction
```

```python
print("{} nn score: {}".format(11,knn.score(x_test,y_test)))
```

```python
# find k value
score_list = []
for each in range(1,15):
    knn2 = KNeighborsClassifier(n_neighbors= each)
    knn2.fit(x_train,y_train)
    score_list.append(knn2.score(x_test,y_test))
    
plt.plot(range(1,15),score_list)
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.show()
```
