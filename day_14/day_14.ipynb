```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("data.csv")
data
```

```python
data.drop(["id", "Unnamed: 32"], axis=1, inplace=True)
data.info()
```

```python
M = data[data.diagnosis == "M"]
B = data[data.diagnosis == "B"]

plt.scatter(M.radius_mean, M.texture_mean, color='r',label="kt", marker='.', alpha=0.3)
plt.scatter(B.radius_mean, B.texture_mean, color='g', label="iyi", marker='.', alpha=0.3)
plt.xlabel("Radius_Mean")
plt.ylabel("Texture_Mean")
plt.legend()
plt.show()
```

```python
y = data.diagnosis.values
x_data = data.drop(["diagnosis"], axis=1)
```

```python
y = [1 if each == 'M' else 0 for each in data.diagnosis]
```

```python
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data)) # normalization
```

```python
x
```

```python
print("y : ", y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=1) # train test split
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train,y_train)
```

```python
print("accuracy of naive bayes algo: ", nb.score(x_test, y_test))
```
